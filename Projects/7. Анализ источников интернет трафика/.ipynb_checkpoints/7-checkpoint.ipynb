### 2.1.1. Рассчитайте DAU, WAU и MAU. Вычислите средние значения этих метрик за весь период. Отобразите изменения метрик во времени на графиках.

```python
df_visit_for_au_metrics = df_visit.copy()

#Для расчета продуктовых(пользовательских) метрик - DAU, WAU, MAU - выделим в отдельные столбцы день, неделю, месяц:
df_visit_for_au_metrics['day'] = df_visit_for_au_metrics['start_ts'].dt.date
df_visit_for_au_metrics['week'] = df_visit_for_au_metrics['start_ts'].astype('datetime64[W]')
df_visit_for_au_metrics['month'] = df_visit_for_au_metrics['start_ts'].astype('datetime64[M]')

#Посчитаем метрики и построим графики:
#DAU:
dau = df_visit_for_au_metrics.groupby('day').agg({'uid':'nunique'}).reset_index()
mean_dau = dau['uid'].mean()
dau.plot(x='day', grid=True, figsize=(15,4), color='green', title='DAU - количество уникальных пользователей в день')

#WAU дату недели ищем как разницу датой по дням и днем недели:
df_visit_for_au_metrics['t_week'] = pd.to_datetime(df_visit_for_au_metrics['start_ts'], unit='d'
) - pd.to_timedelta(df_visit_for_au_metrics['start_ts'].dt.dayofweek, unit='d')

df_visit_for_au_metrics['t_week'] = df_visit_for_au_metrics['t_week'].dt.date
wau = df_visit_for_au_metrics.groupby('t_week').agg({'uid':'nunique'}).reset_index()
mean_wau = wau['uid'].mean()
wau.plot(x='t_week', grid=True, kind='bar', figsize=(15,4), color='darkgrey', title='WAU - количество уникальных пользователей в неделю')


#MAU:
df_visit_for_au_metrics['month'] = df_visit_for_au_metrics['month'].astype('str')
mau = df_visit_for_au_metrics.groupby('month').agg({'uid':'nunique'}).reset_index()
mean_mau = mau['uid'].mean()
mau.plot(x='month', grid='True', kind='bar', figsize=(15,4), title='MAU - количество уникальных пользователей в месяц')

#Выведем средние значения:
display("Среднее значение DAU:",int(mean_dau))
display("Среднее значение WAU:",int(mean_wau))
display("Среднее значение MAU:",int(mean_mau))
```

<div class="alert alert-info"> Вывод: Средние значения аудитории за весь период составили: DAO - 907, WAU - 5825, MAU - 23228. На всех графиках видно, что в летний период наблюдается просадка по дневной, недельной и месячной аудитории сайта. Пик аудитории приходится на месяцы ноябрь и декабрь.

### 2.1.2. Определите, сколько раз за день пользователи в среднем заходят на сайт. Постройте график, отражающий изменения метрики во времени.

```python
#Определим количество пользователей и количество посещений сайта в день:
day_visit=df_visit_for_au_metrics.groupby('day').agg({'uid':['count','nunique']}).reset_index()
day_visit.columns=['date','sessions','users']

#Определим среднее количество посещений сайта в расчете на одного пользователя и построим график:
day_visit['sessions_users'] = day_visit['sessions']/day_visit['users']
display(day_visit.head(2))
day_visit.plot(x='date', y='sessions_users', figsize=(16,3), title='Среднее количество сессий в день на одного пользователя')
plt.show()
```

<div class="alert alert-info"> Вывод: В среднем количество посещений сайта на одного пользователя составляет чуть выше 1 и равномерно на протяжении всего года, за исключением последних чисел декабря, в которых значение увеличивается до 1,2.

### 2.1.3. Исследуйте, сколько времени пользователи проводят на сайте. Узнайте продолжительность типичной пользовательской сессии за весь период. Чтобы выбрать подходящую среднюю меру, постройте график распределения. Изменение продолжительности сессий во времени изучать необязательно.

```python
#Рассчитаем продолжительнось каждой сессии:
df_visit_for_au_metrics['session_lenght'] = (df_visit_for_au_metrics['end_ts']-df_visit_for_au_metrics['start_ts']).dt.seconds

#Построим гистограмму(график распределения), чтобы определить подходящую среднюю меру:
df_visit_for_au_metrics['session_lenght'].hist(bins=100, range=[0,6000], figsize=(16,6))

#Рассчитаем моду:
display(df_visit_for_au_metrics['session_lenght'].mode())
```

### 2.1.4. Рассчитайте Retention Rate, применяя когортный анализ. Покажите изменения метрики во времени на графике. Найдите средний Retention Rate на второй месяц «жизни» когорт.

```python
#Определим событие и дату для формирования когорты - возьмем дату первого посещения сайта:
first_visit = df_visit.groupby(['uid'])['start_ts'].min()
first_visit.name = 'first_visit_date'

#Присоединим дату первого посещения к основной таблице:
df_visit_for_ret = df_visit.join(first_visit, on='uid')

#Выделим месяц из start_ts и first_active_date:
df_visit_for_ret['start_ts_month'] = df_visit_for_ret['start_ts'].astype('datetime64[M]')
df_visit_for_ret['first_visit_month'] = df_visit_for_ret['first_visit_date'].astype('datetime64[M]')

#Расчитаем Lifetime:
df_visit_for_ret['lifetime'] = df_visit_for_ret['start_ts_month'] - df_visit_for_ret['first_visit_month']
df_visit_for_ret['lifetime'] = (df_visit_for_ret['lifetime'] / np.timedelta64(1,'M')).round().astype('int')

#Сгруппируем по месяцу первого посещения и по последующим месяцам посещения(lifetime):
cohorts = df_visit_for_ret.groupby(['first_visit_month','lifetime']).agg({'uid':'nunique'}).reset_index()

#Рассчитаем Retention Rate:
#Определим начальное количество пользователей в когорте:
initial_users_count = cohorts[cohorts['lifetime'] == 0][['first_visit_month', 'uid']]

#Переименуем столбец uid в cohort_users:
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})

#Соединим начальное количество каждой когорты - initial_users_count с данными по когортам - cohorts:
cohorts = cohorts.merge(initial_users_count, on='first_visit_month')

#Посчитаем показатель retention rate:
cohorts['retentions'] = (cohorts['uid'] / cohorts['cohort_users']*100).round(2)

#Для отображения на графике в строках когорт только даты прееобразуем в тип - str:
cohorts['first_visit_month'] = cohorts['first_visit_month'].astype('str')

#Построим сводную таблицу и отобразим тепловую карту:
pivot_retention = cohorts.pivot_table(index='first_visit_month',columns='lifetime',values='retentions',aggfunc='sum')

#Добавим расчет среднего по времени жизни среди всех когорт:
pivot_retention.loc['mean'] = pivot_retention.mean()

#Уберем 0-й lifetime, необходимо для выбора нужного диапазона расцвечивания цветов на heatmap: 
pivot_retention = pivot_retention.iloc[:,1:12]
display(pivot_retention)
plt.figure(figsize=(16,4))
plt.title('Retention Rate')
sns.heatmap(pivot_retention, annot=True, fmt='.2f', linewidths=0.5, linecolor='grey')
plt.show()

#Найдем средний Retention Rate на второй месяц жизни когорт:
mean_retention_rate = pivot_retention.iloc[:,1].mean(axis=0)
display(mean_retention_rate)
```

<div class="alert alert-info"> Вывод: Retention Rate на второй месяц жизни когорт составляет - 4,63%. Пользователи первой когорты(2017-06-01) имеют наименьшую динамику снижения Retention Rate - метрика снизилась с 7,87% до ~4% за 10 месяцев. В то время как последующие когорты снижают Retention Rate до ~2% и каждая последующая когорта дедает это за меньшее количество месяцев.

### Метрики электронной коммерции

### 2.2.1. Исследуйте, сколько времени в среднем проходит с момента первого посещения сайта до совершения покупки. Изучать изменение этой метрики во времени необязательно.

```python
df_visit_copy = df_visit.copy()

#Найдем дату первого посещеня сайта:
first_view_site = df_visit_copy.groupby(['uid'])['start_ts'].min()
first_view_site.name = 'first_view_date'

#Найдем дату первой покупки:
first_order = df_order.groupby(['uid'])['buy_ts'].min()
first_order.name = 'first_order_date'

#Соединим таблицу первого посещения сайта с общей таблицей:
view_and_order = df_visit_copy.join(first_view_site, on='uid')

#Соединим таблицу первой покупки с получившейся таблицей:
view_and_order = view_and_order.join(first_order, on='uid')

#Рассчитаем врямя до покупки:
view_and_order['time_to_buy'] = (view_and_order['first_order_date'] - view_and_order['first_view_date']).dt.seconds
display(view_and_order)

#Рассчитаем - сколько времени в среднем проходит с момента первого посещения сайта до совершения покупки в часах.
display((view_and_order['time_to_buy'].mean()/3600).round(2))
```

<div class="alert alert-info"> Вывод: В среднем 7 часов занимает время с момента первого посещения сайта и до совершения первой покупки.

### 2.2.2. Применяя когортный анализ или другой метод, рассчитайте среднее количество покупок на одного покупателя за определённый период, например за 6 месяцев. Изучать изменение этой метрики во времени необязательно.

```python
df_order_copy = df_order.copy()
#display(df_order.head(1))

#Выделим месяц покупки:
df_order_copy['order_month'] = df_order_copy['buy_ts'].astype('datetime64[M]')
#display(df_order_copy.head(1))

#Получим месяц первой покупки каждого покупателя:
first_order = df_order_copy.groupby('uid').agg({'order_month':'min'}).reset_index()
first_order.columns = ['uid','first_order_month']
#display(first_order.head(1))

#Добавим месяц первой покупки каждого покупателя в таблицу с заказами:
df_order_new = pd.merge(df_order_copy, first_order, on='uid')
#display(df_order_new.head(2))

#Определим lifetime
df_order_new['lifetime'] = df_order_new['order_month'] - df_order_new['first_order_month']
df_order_new['lifetime'] = (df_order_new['lifetime'] / np.timedelta64(1,'M')).round().astype('int')

#Построим когорты, получим количество покупок и количество покупателей когорты:
cohort = df_order_new.groupby(['first_order_month','lifetime']).agg({'uid':['count','nunique']}).reset_index()
cohort.columns = ['first_order_month','lifetime','total_orders','total_buyers']



##НОВОЕ РЕШЕНИЕ:
#Определим начальное количество пользователей в когорте:
initial_users_count = cohort[cohort['lifetime'] == 0][['first_order_month', 'total_buyers']]

#Переименуем столбец total_buyers в cohort_users:
initial_users_count = initial_users_count.rename(columns={'total_buyers':'cohort_users'})

#Соединим начальное количество каждой когорты - initial_users_count с данными по когортам - cohort:
cohort = cohort.merge(initial_users_count, on='first_order_month')

cohort['mean_purchases'] = cohort['total_orders']/cohort['cohort_users']

#Для отображения на графике в строках когорт только даты прееобразуем в тип - str:
cohort['first_order_month'] = cohort['first_order_month'].astype('str')

#Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа:
cohort_lifetime = (cohort.pivot_table(index='first_order_month', columns='lifetime', values='mean_purchases')).round(3)

#Посчитаем средние значения по когоратам с накоплением:
cohort_lifetime = cohort_lifetime.cumsum(axis=1)
##



#Выделим когорты у которых lifetime достиг 6-го месяца и найдем среднее число покупок на одного покупателя за 6 месяцев по когортам:
cohort_lifetime = cohort_lifetime.iloc[0:6,0:7]
#cohort_lifetime['mean_purchases'] = cohort_lifetime.mean(axis=1)
display(cohort_lifetime)

#Построим график:
plt.figure(figsize=(16,5))
plt.title('Срднее количество покупок')
sns.heatmap(cohort_lifetime, annot=True, fmt='.3f', linewidths=0.5, linecolor='grey', vmin=1, vmax=2)
plt.show()
```

<div class="alert alert-info"> Вывод: Наибольшее среднее количество покупок на одного покупателя за 6 месяцев приходится на первую когорту(2017-06-01) и составляет 1,8. Наименьшее среднее количество покупок на одного покупателя за 6 месяцев приходится на когорту(2017-10-01) и составляет - 1,355. По остальным когортам это значение распределенно равномерно и составляет 1,4 - 1,47 покупки.

### 2.2.3. Рассчитайте средний чек, применяя группировку по времени совершения покупки. Когортный анализ не подходит — он может исказить расчёты. Постройте график, отражающий изменения метрики во времени.

### 2.2.4. Выясните, как меняется LTV на покупателя по когортам. Помните, что LTV — накопительная метрика. Рассчитайте средний LTV по когортам за 6 месяцев; в расчёт включайте когорты, «прожившие» не менее 6 месяцев. Маржинальность сервиса — 100%. Отразите изменения метрики во времени на графике.

```python
df_order_ltv = df_order.copy()

#Выделим месяц из таблиц заказов и расходов:
df_order_ltv['month'] = df_order_ltv['buy_ts'].astype('datetime64[M]')

#Вычислим первый месяц покупки:
first_order = df_order_ltv.groupby('uid').agg({'month':'min'}).reset_index()
first_order.columns = ['uid', 'first_order_month']

#Вычислим количество новых покупателей за каждый месяц(когорту):
cohort_sizes = first_order.groupby('first_order_month').agg({'uid':'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'total_buyers']

#Соединим таблицу заказов с месяцем первой покупки:
orders = pd.merge(df_order_ltv, first_order, on='uid')

#Сгруппируем по месяцу и месяцу первой покупки и просуммируем выручку:
cohorts=orders.groupby(['month','first_order_month']).agg({'revenue':'sum'}).reset_index()

#Добавим в таблицу когорт количество покупателей в месяц:
df_cohorts = pd.merge(cohorts, cohort_sizes, on='first_order_month')

#Посчитаем маржтнальность и создадим столбцы с возрастом когорты:
margin = 1
df_cohorts['gross_profit'] = df_cohorts['revenue'] * margin
df_cohorts['cohort_age'] = ((df_cohorts['month'] - df_cohorts['first_order_month'])/np.timedelta64(1, 'M')).round().astype('int')

#Посчитаем LTV:
df_cohorts['ltv'] = df_cohorts['gross_profit']/df_cohorts['total_buyers']
snd = df_cohorts.copy()

#Для отображения в строках когорт только даты прееобразуем в тип - str:
df_cohorts['first_order_month'] = df_cohorts['first_order_month'].astype('str')

#Создадим сводную таблицу по когортам и возрасту когорты:
cohorts_and_ages = (df_cohorts.pivot_table(index='first_order_month', columns='cohort_age', values='ltv', aggfunc='mean')).round(2)
cohorts_and_ages = cohorts_and_ages.cumsum(axis=1)
display(cohorts_and_ages)

#Выделим когорты существующие не менее 6-ти месяцев и найдем среднее значение LTV по этим когортам:
cohorts_and_ages = cohorts_and_ages.iloc[0:6,0:7]
cohorts_and_ages['mean_ltv'] = cohorts_and_ages.mean(axis=1)
display(cohorts_and_ages)

#Построим график изменения среднего значения LTV во времени:
plt.figure(figsize=(16,5))
plt.title('Среднее значение LTV по когортам за 6 месяцев')
sns.heatmap(cohorts_and_ages, annot=True, fmt='.1f', linewidths=0.5, linecolor='grey',  vmin=3, vmax=14)
plt.show()
```

<div class="alert alert-info"> Вывод: Значения LTV находятся в значениях от 5,8 до 9,7. Самое низкое значение LTV - 5.8 приходится на когорту 2017-10-01. Самое высокое значение LTV - 9,7 приходится на когорту 2017-09-01.

### 2.3.1. Посчитайте общую сумму расходов на маркетинг. Выясните, как траты распределены по источникам. Визуализируйте изменения метрик во времени.

```python
df_order_cac = df_order.copy()
df_visit_cac = df_visit.copy()
df_cost_cac = df_cost.copy()

#Выделим месяц из дат в таблице заказов и определим первый месяц заказа каждого пользователя:
df_order_cac['buy_ts'] = df_order_cac['buy_ts'].astype('datetime64[M]')
first_order = df_order_cac.groupby(['uid']).agg({'buy_ts':'min'}).reset_index()
first_order.columns = ['uid', 'first_order_date']

#Создадим таблицу визитов с первыми источниками:
df_visit_cac = df_visit_cac.sort_values('start_ts').groupby('uid').first()

#Объединим таблицы first_order и df_visit_cac:
source_first_visit = pd.merge(first_order, df_visit_cac, on='uid')

#Сгруппируем таблицу визитов по источнику и дате первого заказа и посчитаем количество пользователей:
group_source_first_visit = source_first_visit.groupby(['source_id','first_order_date']).agg({'uid':'nunique'}).reset_index()
group_source_first_visit = group_source_first_visit.rename(columns={'uid':'n_buyers'})

#Определим месяц расхода:
df_cost_cac['dt'] = df_cost_cac['dt'].astype('datetime64[M]')

#Сгруппируем таблицу cost по источнику и дате с суммированием расходов:
group_df_cost_cac = df_cost_cac.groupby(['source_id', 'dt']).agg({'costs':'sum'})

#Объединим таблицы group_source_first_visit и group_df_cost_cac:
cost_by_source = pd.merge(group_source_first_visit, group_df_cost_cac, left_on=['first_order_date','source_id'], right_on=['dt','source_id'])
cost_by_source = cost_by_source.rename(columns={'first_order_date':'month'})
display(cost_by_source)

#Для отображения в строках когорт только даты прееобразуем в тип - str:
cost_by_source['month'] = cost_by_source['month'].astype('str')

#Построим сводную таблицу и график распределения трат по источникам во времени:
pivot = cost_by_source.pivot_table(index='month', columns='source_id', values='costs')
display(pivot)

pivot.plot(style='-o', figsize=(16,5), grid=True, title='Распределения затрат по источникам во времени')
plt.show()
```

<div class="alert alert-info"> Выводы: Самые высокие траты приходятся на источники 2, 3, 4 и 5 и на месяцы с октября по март. Самые низкие значения расходов приходятся на источники 9 и 10 без пиковых значений по времени. Самые высокие траты идут на источник 3.

### 2.3.3. Рассчитайте ROMI по когортам в разрезе источников. Сравните окупаемость за одинаковые периоды жизни когорт. Обратите внимание, что клиенты, пришедшие из разных источников, могут иметь разный LTV. Постройте графики, отражающие изменения метрик во времени.

```python
df_order_ltv = df_order.copy()
df_visit_ltv = df_visit.copy()

#Выделим месяц покупки в таблице заказов:
df_order_ltv['buy_ts'] = df_order_ltv['buy_ts'].astype('datetime64[M]')

##Определим первый месяц покупки:
first_order = df_order_ltv.groupby(['uid']).agg({'buy_ts':'min'}).reset_index()
first_order.columns = ['uid', 'first_order_date']

#Объединим first_order_ltv и first_order:
df_order_ltv = pd.merge(df_order_ltv, first_order, on='uid')

#Выделим месяц покупки в таблице визитов:
df_visit_ltv['start_ts'] = df_visit_ltv['start_ts'].astype('datetime64[M]')

#Создадим таблицу визитов с первыми источниками:
df_visit_ltv = df_visit_ltv.sort_values('start_ts').groupby('uid')['source_id'].first()

#Соединим эти таблицы вместе - df_order_ltv и df_visit_ltv
df_order_visit_ltv = pd.merge(df_order_ltv, df_visit_ltv, on='uid')

#Сгруппируем полученную таблицу по источнику трафика и по дате с подсчетом уникальных пользователей:
group_order_visit_ltv = df_order_visit_ltv.groupby(['first_order_date','source_id']).agg({'uid':'nunique'}).reset_index()
#display(group_order_visit_ltv)

#Считаем revenue по месяцу первой покупки, месяцу покупки и источнику:
group_first_order_date = df_order_visit_ltv.groupby(['first_order_date','buy_ts','source_id']).agg({'revenue':'sum'}).reset_index()
#display(group_first_order_date)

#Соединим таблицы:
df_group_order_visit_source_ltv = pd.merge(group_order_visit_ltv,group_first_order_date, on=['first_order_date','source_id'])
#display(df_group_order_visit_source_ltv)

#Соединим полученную таблицу с посчитанной revenue c общей таблицей из предыдущего раздела с подсчитанными затратами: 
cost_by_source['month'] = cost_by_source['month'].astype('datetime64[M]')
report = pd.merge(df_group_order_visit_source_ltv, cost_by_source, left_on = ['source_id','first_order_date'], right_on = ['source_id','month'])
#display(report)

#Посчитаем LTV:
report['ltv'] = report['revenue'] / report['uid']
#display(report)

#Посчитаем ROMI:
report['romi'] = report['ltv'] / report['cac']
#display(report)


#Добавим возраст когорт:
report['age'] = (report['buy_ts'] - report['first_order_date'])/np.timedelta64(1,'M')
report['age'] = report['age'].round().astype('int')

#Построим сводную таблицу:
pivot = report.pivot_table(index = 'source_id', columns='age', values='romi', aggfunc='mean')

#Рассчитаем накопительный ROMI:
cumulative_romi = pivot.cumsum(axis=1).round(3)
display(cumulative_romi)

#График:
plt.figure(figsize=(16,5))
plt.title('Распределение окупаемости когорт по источникам во времени', loc=('right'))
sns.heatmap(cumulative_romi, annot=True, fmt='.3f', linewidths=0.5, linecolor='grey', vmin=0, vmax=2.4)
plt.show()
```

Вывод:

- Когорта пришедшая по источнику 1 - окупилась на 2 месяц(ROMI=1.032).
- Когорта по источнику 5 на 4 - месяц(ROMI=1.041).
- Когорта по источнику 2 на 5 - месяц(ROMI=1.043).
- Когорта по источнику 9 на 6 - месяц(ROMI=1.009).
- Когорта по источнику 4 на 9 - месяц(ROMI=1.011).
- Когорта по источнику 10 на 11 - месяц(ROMI=1.016).
- Когорта по источнику 3 - не окупилась.

```python
#Рассчитаем накопительный ROMI по каждому источнику:

#создадим список источников:
id_sources = report['source_id'].unique()

#С помощью цикла выведем тепловые карты распределения окупаемости по каждому источнику:
for n in id_sources:
    source = report[report['source_id']==n]
    source['first_order_date'] = source['first_order_date'].astype('str')
    pivot_by_source = source.pivot_table(index='first_order_date', columns='age', values='romi', aggfunc='mean')
    cumulative_romi = pivot_by_source.cumsum(axis=1).round(3)

    plt.figure(figsize=(16,5))
    plt.title(n, loc=('right'))
    sns.heatmap(cumulative_romi, annot=True, fmt='.3f', linewidths=0.5, linecolor='grey')
    plt.show()
```
