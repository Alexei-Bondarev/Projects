<h2>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ V.2</h2>

–ù—É –≤ —ç—Ç–æ–º —Å–º—ã—Å–ª–µ —Ç—ã –ø—Ä–∞–≤ :) –ù–æ –µ—Å–ª–∏ —ç—Ç–æ –æ—Ç–±—Ä–æ—Å–∏—Ç—å, —Ç–æ –≤ –∫–∞–∫–∏—Ö-—Ç–æ —Ç–∏–ø–∞—Ö –≤–µ—Ä–æ—è—Ç–Ω–µ–µ –Ω–∞—Ç–∫–Ω—É—Ç—å—Å—è –Ω–∞ —Å–µ—Ç–µ–≤–æ–µ –∑–∞–≤–µ–¥–µ–Ω–∏–µ, —á–µ–º –≤ –¥—Ä—É–≥–∏—Ö:
</div>

```python
# –ö–æ–¥ —Ä–µ–≤—å—é–µ—Ä–∞
chain_by_rest_type = data.groupby('object_type').apply(lambda x: pd.Series(dict(
    chained=(x.chain == '—Å–µ—Ç–µ–≤—ã–µ').sum(),
    not_chained=(x.chain == '–Ω–µ—Å–µ—Ç–µ–≤—ã–µ').sum(),
    total =x.id.count()
)))


chain_by_rest_type = chain_by_rest_type.reset_index()
chain_by_rest_type['chained_share'] = chain_by_rest_type['chained']/chain_by_rest_type['total']*100
chain_by_rest_type['not_chained_share'] = chain_by_rest_type['not_chained']/chain_by_rest_type['total']*100

chain_by_rest_type.sort_values(by='chained_share', ascending=False)
```

### –í—ã–¥–µ–ª–∏—Ç–µ –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± —É–ª–∏—Ü–µ –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ address.

```python
# –° –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞–∑–¥–µ–ª–∏–º –∑–Ω–∞—á–µ–Ω–∏—è –∞–¥—Ä–µ—Å–∞ –≤ —Å—Ç–æ–ª–±—Ü–µ - address, –Ω–∞ —ç–ª–µ–º–µ–Ω—Ç—ã, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã–µ –∑–∞–ø—è—Ç–æ–π:
data['split_street'] = None
def split_street(string):
    street = string.split(',')
    return street

data['split_street'] = data['address'].apply(split_street)

# –°–ø–∏—Å–æ–∫ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–π —Ç–∏–ø–æ–≤ —É–ª–∏—Ü:
mask = ['—É–ª–∏—Ü–∞', '–±—É–ª—å–≤–∞—Ä', '—à–æ—Å—Å–µ','–ø–µ—Ä–µ—É–ª–æ–∫','–ø–µ—Ä–µ—É–ª–æ–∫', '–ø—Ä–æ–µ–∑–¥', '–ø—Ä–æ—Å–ø–µ–∫—Ç', '–Ω–∞–±–µ—Ä–µ–∂–Ω–∞—è', '–ø–ª–æ—â–∞–¥—å', '–∞–ª–ª–µ—è', '–∫–≤–∞—Ä—Ç–∞–ª', '–ü—Ä–æ–µ–∑–¥'
       , '–º–∏–∫—Ä–æ—Ä–∞–π–æ–Ω', '–≥–æ—Ä–æ–¥ –ó–µ–ª–µ–Ω–æ–≥—Ä–∞–¥', '–¥–µ—Ä–µ–≤–Ω—è', '–ø–æ—Å—ë–ª–æ–∫', '—Å–µ–ª–æ', '–ª–∏–Ω–∏—è', '—Ç—É–ø–∏–∫', '87-–π –∫–∏–ª–æ–º–µ—Ç—Ä', '80-–π –∫–∏–ª–æ–º–µ—Ç—Ä'
        , '82-–π –∫–∏–ª–æ–º–µ—Ç—Ä', '33-–π –∫–∏–ª–æ–º–µ—Ç—Ä', '38-–π –∫–∏–ª–æ–º–µ—Ç—Ä', '104-–π –∫–∏–ª–æ–º–µ—Ç—Ä', '19-–π –∫–∏–ª–æ–º–µ—Ç—Ä', '73-–π –∫–∏–ª–æ–º–µ—Ç—Ä'
        ,'—Ç–µ—Ä—Ä–∏—Ç–æ—Ä–∏—è –õ–µ–Ω–∏–Ω—Å–∫–∏–µ –ì–æ—Ä—ã', '—Ç–µ—Ä—Ä–∏—Ç–æ—Ä–∏—è –°–ù–¢ "–ì–∞–≤—Ä–∏–∫–æ–≤–æ-1"']

# –° –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ –≤—ã–¥–µ–ª–∏–º –∏–∑ —Å–ø–∏—Å–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤(—Å—Ç–æ–ª–±–µ—Ü split street) –Ω–∞–∑–≤–∞–Ω–∏–µ —É–ª–∏—Ü—ã:
def find_street(i):
    for k in range(len(i)):
        for m in mask:
            if m in i[k]:
                return i[k]
            
data['street'] = data['split_street'].apply(find_street)
display(data)
```

<div class="alert alert-info"> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞: –¢–∞–∫ –∫–∞–∫ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–¥–∞–Ω–∏–π –ø—Ä–æ–µ–∫—Ç–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã —Ä–∞–π–æ–Ω—ã, –≤ –∫–æ—Ç–æ—Ä—ã—Ö —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω—ã —É–ª–∏—Ü—ã, –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Å–∞–π—Ç–æ–º - http://mosopen.ru/streets, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º –µ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ–± —É–ª–∏—Ü–∞—Ö –≥. –ú–æ—Å–∫–≤—ã —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –ø–æ —Ä–∞–π–æ–Ω–∞–º. –°—Ñ–æ—Ä–º–∏—Ä—É–µ–º –Ω–æ–≤—ã–π DataFrame, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å –ø–∞—Ä—ã –∑–Ω–∞—á–µ–Ω–∏–π - —É–ª–∏—Ü–∞ - —Ä–∞–π–æ–Ω—ã –≤ –¥–≤–∞ —ç—Ç–∞–ø–∞.

<div class="alert alert-success">
<h2>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞</h2>

–•–æ—Ä–æ—à–∞ –∏–¥–µ—è –∏ —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ –∞–∫–∫—É—Ä–∞—Ç–Ω–æ üëç –ú–æ–∂–Ω–æ –±—ã–ª–æ –Ω–∞–≤–µ—Ä–Ω–æ–µ –æ–±–æ–π—Ç–∏—Å—å –æ–¥–Ω–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–º –∏ —Ä—É–∫–∞–º–∏ –Ω–µ –¥—Ä–æ–ø–∞—Ç—å –Ω–µ—É–¥–æ–±–Ω—ã–µ —Ä–∞–π–æ–Ω—ã, –∞ –≤ –∫–∞–∫–æ–π-—Ç–æ –º–æ–º–µ–Ω—Ç –ª–æ–≤–∏—Ç—å –∏—Å–∫–ª–±—á–µ–Ω–∏–µ, –µ—Å–ª–∏ —á—Ç–æ-—Ç–æ –Ω–µ –Ω–∞—à–ª–æ—Å—å, –Ω–æ —ç—Ç–æ –º–µ–ª–æ—á–∏
</div>

```python
# –≠—Ç–∞–ø 1:

# –ü–æ–ª—É—á–∏–º —Ç–∞–±–ª–∏—Ü—É —Å–æ–¥–µ—Ä–∂–∞—â—É—é - –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Ä–∞–π–æ–Ω–∞ –∏ —Å—Å—ã–ª–∫—É –≤–µ–¥—É—â—É—é –Ω–∞ —Å—Ç—Ä–∞–Ω–∏—Ü—É —Å–æ —Å–ø–∏—Å–∫–æ–º —É–ª–∏—Ü —Ä–∞–π–æ–Ω–∞:
URL = 'http://mosopen.ru/streets'
req = requests.get(URL)
soup = BeautifulSoup(req.text, 'lxml')

# –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç—ã–µ —Å–ø–∏—Å–∫–∏:
district_ref_list = []
district_name_list = []

# –ó–∞–ø–æ–ª–Ω—è–µ–º —Ç–∞–±–ª–∏—Ü—É –∑–∞–ø–∏—Å—è–º–∏(—Å—Å—ã–ª–∫–∞ - —Ä–∞–π–æ–Ω), –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ —Ç–µ–≥–µ <a> —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º 'href', —Å–æ–¥–µ—Ä–∂–∞—â–∏–º —Å—Å—ã–ª–∫—É –≤–∏–¥–∞ - "^http://mosopen.ru/region/"
for row in soup.find_all('a', attrs={'href': re.compile("^http://mosopen.ru/region/")}):
    dist_ref = row.get('href')
    dist_name = row.text   
    district_ref_list.append(dist_ref)
    district_name_list.append(dist_name)

district_data = {'district_ref' : district_ref_list, 'district_name' : district_name_list}
d_data = pd.DataFrame(district_data)

# –£–¥–∞–ª–∏–º –¥—É–±–ª–∏–∫–∞—Ç—ã –∏–∑ —Ç–∞–±–ª–∏—Ü—ã:
d_data.drop_duplicates(inplace=True)

# –£–¥–∞–ª–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º–∏ —Å—Ç—Ä–∞–Ω–∏—Ü–∞–º–∏:
d_data = d_data.drop(d_data.index[20]).reset_index(drop=True)
d_data = d_data.drop(d_data.index[69]).reset_index(drop=True)
d_data = d_data.drop(d_data.index[72])
display(d_data.head(3))
```

```python
# –≠—Ç–∞–ø 2:

pd.set_option('display.max_rows', 4180)

# –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π DataFrame:
all_street_list = pd.DataFrame()


for path, name in zip(d_data['district_ref'], d_data['district_name']):    
    URL = path
    req = requests.get(URL)
    soup = BeautifulSoup(req.text, 'lxml')
    street_list = []
    for row in soup.find_all('div', attrs={'class': 'double_block clearfix'}):
        street_list.append(row.text)

# –†–∞–∑–¥–µ–ª–∏–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é - '\n':
    st = street_list[0].split('\n')

# –£–¥–∞–ª–∏–º –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è - ''
    s_list = [st for st in st if st]

# –ü–æ–º–µ—Å—Ç–∏–º –∑–Ω–∞—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ –≤ DataFrame:
    s_data = pd.DataFrame(s_list)

#–°–æ–∑–¥–∞–¥–∏–º —Å—Ç–æ–ª–±–µ—Ü –¥–ª—è –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è —Ä–∞–π–æ–Ω–æ–≤:
    s_data['dist_name'] =  name
    
#–°–æ–±–∏—Ä–∞–µ–º —É–ª–∏—Ü—ã –∫–∞–∂–¥–æ–≥–æ —Ä–∞–π–æ–Ω–∞ –≤ –æ–¥–∏–Ω DataFrame:    
    all_street_list = pd.concat([all_street_list, s_data])           

# –ü–µ—Ä–µ–∏–º–µ–Ω—É–µ–º —Å—Ç–æ–ª–±—Ü—ã:
all_street_list.columns = ['street_name', 'dist_name']
display(all_street_list.head(2))
```

```python
# –î–æ–±–∞–≤–∏–º —Ä–∞–π–æ–Ω—ã –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–æ—Ö–æ–¥–∏—Ç —É–ª–∏—Ü–∞ –≤ –∑–∞–ø–∏—Å—å —É–ª–∏—Ü—ã:
group_street = all_street_list.copy()
group_street = group_street.groupby('street_name')['dist_name'].apply(lambda x: x.unique()).reset_index()

display(group_street.head(2))
```

```python
#–£–¥–∞–ª–∏–º —Å–∏–º–≤–æ–ª—ã –ø—Ä–æ–±–µ–ª–∞ –≤ –∞–¥—Ä–µ—Å–∞—Ö –ø–æ –∫—Ä–∞—è–º —Å—Ç—Ä–æ–∫–∏ –≤ –¥–≤—É—Ö —Ç–∞–±–ª–∏—Ü–∞—Ö:
data['street'] = data['street'].str.strip()
group_street['street_name'] = group_street['street_name'].str.strip()

# –£–¥–∞–ª–∏–º –∑–∞–ø—è—Ç—ã–µ –∏–∑ –∞–¥—Ä–µ—Å–æ–≤:
data['street'] = data['street'].str.replace(',','')
group_street['street_name'] = group_street['street_name'].str.replace(',','')

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —É–ª–∏—Ü, —Å–æ–¥–µ—Ä–∂–∞—â–∏—Ö –∑–∞–ø—è—Ç—ã–µ:
display(data[data['street'].str.contains(',')])
display(group_street[group_street['street_name'].str.contains(',')])
```

```python
# –¢–∞–∫ –∫–∞–∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Ä–∞–π–æ–Ω–æ–≤ —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —É–ª–∏—Ü –≤ —Ñ–æ—Ä–º–∞—Ç–µ: "–Ω–∞–∑–≤–∞–Ω–∏–µ —É–ª–∏—Ü–∞", –∞ –≤ –∏—Å—Ö–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è —Ä–∞–∑–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã:
# –∏ "–Ω–∞–∑–≤–∞–Ω–∏–µ —É–ª–∏—Ü–∞" –∏ "—É–ª–∏—Ü–∞ –Ω–∞–∑–≤–∞–Ω–∏–µ" —Ç–æ –≤ –∏—Å—Ö–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ –ø–µ—Ä–µ–Ω–µ—Å–µ–º —Å–ª–æ–≤–æ —É–ª–∏—Ü–∞ –≤ –∫–æ–Ω–µ—Ü —Å—Ç—Ä–æ–∫–∏ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è —É–ª–∏—Ü—ã:

def format_line(i):
    for m in mask:
        if i.startswith(m):            
            i=i.replace(m,'')
            i = i + ' ' + m            
            return i
    return i

# –ü—Ä–æ–≤–µ—Ä–∏–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Ä–µ—É–ª—å—Ç–∞—Ç –≤ —Å—Ç–æ–ª–±—Ü–µ - format_street:
data['format_street'] = data['street'].apply(format_line)
display(data)
```
